```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split

df = pd.read_csv("/content/iris.csv")

df['species'].replace({'setosa':0, 'versicolor':1, 'virginica':2}, inplace = True)

x = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df['species']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2 , random_state= 11)

df = DecisionTreeClassifier(random_state = 11)
df.fit(x_train, y_train)

pred = df.predict(x_test)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(round(acc,2))
print(round(100*acc,1),"%")
```

```python
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)
```

```python
# 비지도학습
# 군집 분석
# k- means 알고리즘 (유사 데이터 동일 그룹 묶기)
# k 클러스터 ,means 평균 거리
# 거리차이 분산 최소화

import numpy as np
import pandas as pd
import sklearn

from sklearn.cluster import KMeans

df = pd.read_csv('/content/iris.csv')

df.info() # species 문자형

# 데이터 전처리
# 문자형 - 숫자형으로 변경
from sklearn.preprocessing import LabelEncoder
df['species'] = LabelEncoder().fit_transform(df['species'])
df.head()

# 분석 데이터셋 준비
x = df[['sepal_length','sepal_width','petal_length','petal_width']]

# KMeans
cluster1 = KMeans(n_clusters = 3, n_init = 10 , max_iter = 500, random_state = 42)
cluster1.fit(x)

# 군집의 중심점, 예측 군집 결과값
cluster_centers = cluster1.cluster_centers_
cluster_prediction = cluster1.predict(x)

print(pd.DataFrame(cluster_centers))
print(cluster_prediction)
df['cluster'] = cluster_prediction
df

# 성능평가
# 정답이 없음
# inertias가 빠르게 줄어드는 지점의 k 값

scope = range(1,10)
inertias = []

for k in scope :
    model =  KMeans(n_clusters = k, n_init = 10 , max_iter = 500 , random_state= 42)
    model.fit(x)
    inertias.append(model.inertia_)
    print(k, inertias[k-1])



```

```python
# 연관 분석
# 관련성 파악
# 비교 규칙 필요 ( 신뢰도 지지도 향상도 )
# support
# confidence
# lift

import numpy as np
import pandas as pd
import sklearn

from mlxtend.frequent_patterns import apriori, association_rules

df = pd.read_csv('/content/retail_dataset.csv',sep=',')
df.info()

# 데이터 전처리

items = set()

for col in df:
    items.update(df[col].unique())

itemset = set(items)
encoding = []

for index , row in df.iterrows():
    rowset = set(row)
    labels = {}
    dismatching = list(itemset - rowset)
    matching = list(itemset.intersection(rowset))

    for i in dismatching:
        labels[i] = 0
    for j in matching:
        labels[j] = 1
    encoding.append(labels)
encoding[0]
result = pd.DataFrame(encoding)

result

# 분석 데이터셋

result = result.drop(result.columns[0], axis = 1)
result

# 연관 분석 apriori 알고리즘 사용
freq_items = apriori(result, min_support = 0.2 , use_colnames = True)
# association_rules 함수로 규칙 도출
rules = association_rules(freq_items, metric = 'confidence', min_threshold = 0.6)
rules.head()
```
