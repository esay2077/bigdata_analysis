```python
"""
데이터 전처리
문자 - 숫자
월 - 1 수치 -정규화
동일 대상 분포 환산

결측치
삭제
대체
NaN

틀린값
버림
대체
분석보류

변환
범주형
일반 정규화
z-score 정규화 평균 0 표준편차 1
로그변환 로그정규분포
역수변환

"""
```

```python
#표준 정규화
meat_buy_kr = 5*np.random.randn(1000) +53.9
meat_buy_jp = 4*np.random.randn(1000) +32.7

meat_buy = pd.DataFrame({"한국인":meat_buy_kr,"일본인":meat_buy_jp})

meat_buy.head()
```

```python
plt.hist(meat_buy_kr)
plt.xlabel('korea')
plt.show()

plt.hist(meat_buy_jp)
plt.xlabel('japan')
plt.show()
```

```python
#z-표준화
df1 = pd.read_csv("/content/df_sample.csv")

mid_avg = np.mean(df1["중간"]) #평균
mid_std = np.std(df1["중간"]) #표준편차

df1["중간_z점수"] = (df1["중간"]-mid_avg) / mid_std

df1.head()
```

```python
import scipy.stats as ss

meat_buy["한국인 정규화"] = ss.zscore(meat_buy_kr)
meat_buy["일본인 정규화"] = ss.zscore(meat_buy_jp)

meat_buy.head()
```

```python
meat_buy["한국인 정규화2"] = (meat_buy_kr - np.mean(meat_buy_kr))/np.std(meat_buy_kr)
meat_buy["일본인 정규화2"] = (meat_buy_jp - np.mean(meat_buy_jp))/np.std(meat_buy_jp)

meat_buy.head()
```

```python
plt.hist(meat_buy['한국인 정규화'])
plt.xlabel('korea')
plt.show()

plt.hist(meat_buy['일본인 정규화'])
plt.xlabel('japen')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
meat_buy['한국인 정규화3'] = scaler.fit_transform(meat_buy[['한국인']])

scaler = StandardScaler()
meat_buy['일본인 정규화3'] = scaler.fit_transform(meat_buy[['일본인']])
```

```python
#정규분포변환
"""
왜도란 대칭정도
양수 : 평균>중앙>최빈
0 : 평 = 중 = 최빈
음수 : 평 < 중 < 최빈
```

```python
df = pd.read_csv("/content/USJudgeRatings.csv")
df.head(5)
```

```python
import scipy.stats as ss

print(ss.skew(df['CONT'])) #양수
print(ss.skew(df["PHYS"])) #음수
```

```python
# 왜도 양수: sqrt log 1/x
# 왜도 음수: sqrt(max(x+1)-x) log(max(x+1)-x) 1/(max(x+1)-x)
```

```python
df['CONT1'] = np.log(df["CONT"])
df['PHYS1'] = np.log(np.max(df['PHYS']+1) -df['PHYS'])

print(ss.skew(df['CONT1']))
print(ss.skew(df["PHYS1"]))
```

```python
df['CONT2'] = np.log10(df["CONT"])
df['PHYS2'] = np.log10(np.max(df['PHYS']+1) -df['PHYS'])

print(ss.skew(df['CONT2']))
print(ss.skew(df["PHYS2"]))
```

```python
#범주화

data = [["철수",52], ["영희",92], ["미영",84], ["시완",71], ["미경",65], ["영환",81], ["숙경",66], ["부영",77], ["민섭",73], ["보연",74]]
df = pd.DataFrame(data,columns=['이름','수학점수'])

print(np.mean(df['수학점수']))

plt.hist(df['수학점수'],bins=5, range=[50,100], rwidth=0.9)
plt.show()
df
```

```python
df['등급'] = 0

df.loc[(df['수학점수']<60),'등급'] = 'F'
df.loc[(df['수학점수']>=60) & (df['수학점수']<=70),'등급'] = 'D'
df.loc[(df['수학점수']>=70) & (df['수학점수']<=80),'등급'] = 'C'
df.loc[(df['수학점수']>=80)& (df['수학점수']<=90),'등급'] = 'B'
df.loc[(df['수학점수']>=90)& (df['수학점수']<=100),'등급'] = "A"

df
```

```python
df['등급'] = pd.cut(x=df['수학점수'],
                  bins = [0,60,70,80,90,100],
                  labels = ["F","D",'C','B','A'],
                  include_lowest = True)
df
#cut()
```

```python
#qcut()
df['등급_qcut'] = pd.qcut(x=df['수학점수'],q=5, labels = ['F','D','C','B','A'])

df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 정규 분포 평균 50 표준편차 10 데이터 200개
data = 10 * np.random.randn(200) + 50
df = pd.DataFrame({'값':data}) # 데이터 프레임 변환

# 이상치 삽입
df.loc[201] = 2
df.loc[202] = 100
df.loc[203] = 10
df.loc[204] = 110

#시각화
plt.hist(df['값'],bins=20,rwidth=0.8)
plt.show()
```

```python
# IQR 방법
plt.boxplot(df['값'])
plt.show()
```

```python
Q1 = df['값'].quantile(.25)
Q2 = df['값'].quantile(.5)
Q3 = df['값'].quantile(.75)
IQR = Q3 - Q1

print("Q1=", Q1)
print('Q2=', Q2)
print("Q3=",Q3)
print("IQR = ", IQR)
```

```python
Q1 = df['값'].describe()['25%']
Q2 = df['값'].describe()['50%']
Q3 = df["값"].describe()['75%']
IQR = Q3 - Q1

print("Q1 = ",Q1)
print("Q2= ",Q2)
print("Q3= ", Q3)
print("IQR=",IQR)
```

```python
#이상치 검출
condition = df['값'] > (Q3 + IQR * 1.5)
upperOutlier = df[condition]

condition = df['값'] < (Q1 - IQR * 1.5)
lowerOutlier = df[condition]

print(lowerOutlier)
print(upperOutlier)
```
