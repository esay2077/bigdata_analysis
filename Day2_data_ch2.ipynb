```python
#z-í‘œì¤€í™”
df1 = pd.read_csv("/content/df_sample.csv")

mid_avg = np.mean(df1["ì¤‘ê°„"]) #í‰ê· 
mid_std = np.std(df1["ì¤‘ê°„"]) #í‘œì¤€íŽ¸ì°¨

df1["ì¤‘ê°„_zì ìˆ˜"] = (df1["ì¤‘ê°„"]-mid_avg) / mid_std

df1.head()
```

```python
import scipy.stats as ss

meat_buy["í•œêµ­ì¸ ì •ê·œí™”"] = ss.zscore(meat_buy_kr)
meat_buy["ì¼ë³¸ì¸ ì •ê·œí™”"] = ss.zscore(meat_buy_jp)

meat_buy.head()
```

```python
meat_buy["í•œêµ­ì¸ ì •ê·œí™”2"] = (meat_buy_kr - np.mean(meat_buy_kr))/np.std(meat_buy_kr)
meat_buy["ì¼ë³¸ì¸ ì •ê·œí™”2"] = (meat_buy_jp - np.mean(meat_buy_jp))/np.std(meat_buy_jp)

meat_buy.head()
```

```python
import scipy.stats as ss

print(ss.skew(df['CONT'])) #ì–‘ìˆ˜
print(ss.skew(df["PHYS"])) #ìŒìˆ˜
```

```python
# ì™œë„ ì–‘ìˆ˜: sqrt log 1/x
# ì™œë„ ìŒìˆ˜: sqrt(max(x+1)-x) log(max(x+1)-x) 1/(max(x+1)-x)
```

```python
df['CONT1'] = np.log(df["CONT"])
df['PHYS1'] = np.log(np.max(df['PHYS']+1) -df['PHYS'])

print(ss.skew(df['CONT1']))
print(ss.skew(df["PHYS1"]))
```

```python
df['CONT2'] = np.log10(df["CONT"])
df['PHYS2'] = np.log10(np.max(df['PHYS']+1) -df['PHYS'])

print(ss.skew(df['CONT2']))
print(ss.skew(df["PHYS2"]))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ì •ê·œ ë¶„í¬ í‰ê·  50 í‘œì¤€íŽ¸ì°¨ 10 ë°ì´í„° 200ê°œ
data = 10 * np.random.randn(200) + 50
df = pd.DataFrame({'ê°’':data}) # ë°ì´í„° í”„ë ˆìž„ ë³€í™˜

# ì´ìƒì¹˜ ì‚½ìž…
df.loc[201] = 2
df.loc[202] = 100
df.loc[203] = 10
df.loc[204] = 110

#ì‹œê°í™”
plt.hist(df['ê°’'],bins=20,rwidth=0.8)
plt.show()
```

```python
# IQR ë°©ë²•
plt.boxplot(df['ê°’'])
plt.show()
```

```python
Q1 = df['ê°’'].quantile(.25)
Q2 = df['ê°’'].quantile(.5)
Q3 = df['ê°’'].quantile(.75)
IQR = Q3 - Q1

print("Q1=", Q1)
print('Q2=', Q2)
print("Q3=",Q3)
print("IQR = ", IQR)
```

```python
Q1 = df['ê°’'].describe()['25%']
Q2 = df['ê°’'].describe()['50%']
Q3 = df["ê°’"].describe()['75%']
IQR = Q3 - Q1

print("Q1 = ",Q1)
print("Q2= ",Q2)
print("Q3= ", Q3)
print("IQR=",IQR)
```

```python
#ì´ìƒì¹˜ ê²€ì¶œ
condition = df['ê°’'] > (Q3 + IQR * 1.5)
upperOutlier = df[condition]

condition = df['ê°’'] < (Q1 - IQR * 1.5)
lowerOutlier = df[condition]

print(lowerOutlier)
print(upperOutlier)
```
